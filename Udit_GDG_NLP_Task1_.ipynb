## Stemming

**What is Stemming?** <br>
Ans:
Stemming is a process used in language processing that cuts down words to their root or base form. Think of it as a way to simplify words by removing prefixes or suffixes, so different forms of a word can be treated as the same thing.

For example, the words "running," "runner," and "ran" all come from the base word "run." Stemming would strip down these variations to "run," helping computers understand that they're related. This is especially useful in search engines or text analysis, where you want to find all mentions of a concept regardless of the specific word form used.

```python
# Create a word corpus from the sentences using word tokenization and stemming. Also, remember to avoid adding stopwords to your corpus
# Add your code here
from nltk.tokenize import word_tokenize
corpus = []
stop_words = set(stopwords.words('english'))
for sentence in up_sen:
    words = nltk.word_tokenize(sentence)
    stemmed_words = [stemmer.stem(word) for word in words if word not in stop_words and word.isalnum()]
    corpus.extend(stemmed_words)
```

## Lemmatization

**What is Lemmatization?** <br>
Ans: Lemmatization is similar to stemming but a bit more sophisticated. It reduces words to their base or dictionary form, called the lemma, but unlike stemming, it does this in a way that considers the context and the meaning of the word.

For example, lemmatization would take "running" and convert it to "run," just like stemming. However, it also knows that "better" should be lemmatized to "good" based on the context, whereas stemming would not make that connection.

```python
# Create a second word corpus from the sentences using word tokenization and lemmatization. Also, remember to avoid adding stopwords to your new corpus
# Add your code here

corp2 = []

for sentence in up_sen:

    words = word_tokenize(sentence)

    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.lower() not in stop_words]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    corp2.extend(words)
```

**What is the difference between stemming and lemmatization? Is one better than the other? If yes, justify your answer. If no, determine the usecases where these can be used.** <br>
Ans:

### Stemming vs. Lemmatization:

1. **Stemming** is like taking a shortcut. It chops off the ends of words to get to the root. Its quick but not always cleansometimes you end up with words that dont make sense. For example, "fishing" might turn into "fish," which is fine, but "happily" might turn into "happi," which isnt a real word.

2. **Lemmatization** takes the scenic route. It looks at the word, checks its context, and then converts it into its proper base form. So "fishing" would still turn into "fish," but "better" would correctly turn into "good." Its slower, but the result is always a real word.

### Which One is Better?

It depends on what youre doing:

- **Stemming** is great when you just need speed and dont care if the words get a little messy. Think of it like a quick search toolit gets the job done fast but might miss some details.

- **Lemmatization** is the go-to when accuracy matters. If youre doing something where meaning is important, like analyzing text or training a language model, lemmatization gives you cleaner, more accurate results.

```python
# Include all steps of removal of unnecessary characters, lowering, word tokenization, lemmatization and stopwords removal in one function
# Your final corpus should be a list of sentences, not a list of list of words (Why?)
# Add your code here

def preprocess_text(text):
    text = re.sub(r"[^a-zA-Z0-9 ]", "", text)
    text = text.lower()
    tokens = word_tokenize(text)

    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    sentences = " ".join(tokens)

    return sentences


corp = []
preprocessed_sentences = preprocess_text(par)
corp.append(preprocessed_sentences)
print(corp)
```

```python
# Now to play around with this:
def euclidean_distance(a_tf_idf_counter, b_tf_idf_counter) -> float:
    a_vec_sorted_keys = sorted(a_tf_idf_counter)
    b_vec_sorted_keys = sorted(b_tf_idf_counter)

    a_vec = [a_tf_idf_counter[key] for key in a_vec_sorted_keys]
    b_vec = [b_tf_idf_counter[key] for key in b_vec_sorted_keys]

    dist = 0
    for a_i, b_i in zip(a_vec, b_vec):
        dist += (b_i - a_i) ** 2
    return dist ** 0.5

# a (dot) b = abs(a) * abs(b) * cos(theta)
# cosine similarity is defined as a (dot) b / abs(a) * abs(b)
def cosine_similarity(a_tf_idf_counter, b_tf_idf_counter) -> float:
    a_vec_sorted_keys = sorted(a_tf_idf_counter)
    b_vec_sorted_keys = sorted(b_tf_idf_counter)

    a_vec = [a_tf_idf_counter[key] for key in a_vec_sorted_keys]
    b_vec = [b_tf_idf_counter[key] for key in b_vec_sorted_keys]

    dot_product = 0
    abs_a = 0
    abs_b = 0

    for a_i, b_i in zip(a_vec, b_vec):
        dot_product += a_i * b_i

        abs_a += a_i * a_i
        abs_b += b_i * b_i

    return dot_product / (abs_a ** 0.5 * abs_b ** 0.5)
```

**How does the TF-IDF Vectorizer work?** <br>
Ans: The **TF-IDF Vectorizer** is a tool that converts text (like documents or sentences) into numerical data that can be used in machine learning models. It uses the **TF-IDF** (Term Frequency-Inverse Document Frequency) method to evaluate how important each word is in a document relative to the entire corpus.

Heres how it works step by step:

1. **Calculate Term Frequency (TF)**:  
   It first calculates how often a word appears in a document (Term Frequency). For example, if the word "cat" appears 3 times in a document with 10 total words, the TF for "cat" is 3/10.

2. **Calculate Inverse Document Frequency (IDF)**:  
   Next, it calculates how rare or common the word is across all documents (Inverse Document Frequency). Words that appear in many documents will have a lower IDF score, while words that appear in fewer documents will have a higher IDF score.

3. **Combine TF and IDF**:  
   The TF and IDF scores are multiplied together to get the **TF-IDF** score for each word. This helps emphasize words that are important to a specific document but not common across all documents.

4. **Create Vectors**:  
   For each document, the vectorizer creates a vector where each element represents the TF-IDF score of a word in the document. Each word in the vocabulary of the corpus has its own position in this vector.

Research how the Bidirectional Encoder Representations from Transformers (BERT) Model creates Word Embeddings and how it maintains context and captures semantic meaning between words and sentences.

BERT (Bidirectional Encoder Representations from Transformers) is a model that helps computers understand the meaning of words by looking at the whole sentence theyre in. Instead of giving each word a fixed meaning, like older models, BERT figures out what a word means based on the context around it.

For example, take the word bank:
- In I went to the bank to deposit money, BERT knows "bank" refers to a financial institution.
- In The boat was near the riverbank, BERT understands "bank" means the side of a river.

It does this by reading the sentence both forward and backward at the same time, which is why its called **bidirectional**. This helps it catch the full meaning of a word and how it connects to the rest of the sentence.

BERT uses a special method called transformers to process the sentence. Transformers allow BERT to figure out how all the words relate to each other, even if theyre far apart in the sentence. For instance, in The book, which was lying on the table, belongs to Sarah, BERT knows "book" and "belongs to Sarah" are connected, even with other words in between.

In simple terms, BERT is like a smart reader that doesnt just see wordsit understands their meaning based on the full sentence. This makes it really good at tasks like answering questions, summarizing text, and more.
